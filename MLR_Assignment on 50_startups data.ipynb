#calculating VIF

```python
rsq_ad=smf.ols('Administration~Marketing+Research',data=start).fit().rsquared
vif_ad=1/(1-rsq_ad)
rsq_mr=smf.ols('Marketing~Administration+Research',data=start).fit().rsquared
vif_mr=1/(1-rsq_mr)
rsq_re=smf.ols('Research~Administration+Marketing',data=start).fit().rsquared
vif_re=1/(1-rsq_re)
```

```python
star={'Variables':['ad','mr','re'],'vif':[vif_ad,vif_mr,vif_re]}
vif_frame=pd.DataFrame(star)
vif_frame
```

```python
def get_standardized_values(vals):
  return(vals-vals.mean()/vals.std())
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))
plt.title('Residual plot')
plt.xlabel('fittedvalues')
plt.axhline(y=0,color='black',linestyle='--',linewidth=1)
plt.ylabel('residvalues')
plt.show()
```

#Detecting outliers

#finding cooks distance for each observation

```python
import numpy as np
influence=model.get_influence()
(c,_)=influence.cooks_distance
(c,_)
```

```python
(c,_)
```

#plotting cooks distance by using stemplot

```python
fig=plt.subplots(figsize=(15,5))
plt.stem(np.arange(len(start)), np.round(c,3))
plt.xlabel('row index')
plt.ylabel('cooks distance')
plt.title('stem plot')
plt.show()
```
